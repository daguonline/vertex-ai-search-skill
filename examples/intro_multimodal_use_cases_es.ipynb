### Establecer la información del proyecto de Google Cloud y crear el cliente

### Importar librerías


```python
from IPython.display import Audio, Image, Markdown, Video, display
from gitingest import ingest
from google.genai.types import CreateCachedContentConfig, GenerateContentConfig, Part
import nest_asyncio

nest_asyncio.apply()
```

### Cargar el modelo Gemini 2.5 Flash

Obtén más información sobre todos los [modelos Gemini en Vertex AI](https://cloud.google.com/vertex-ai/generative-ai/docs/learn/models#gemini-models).

```python
MODEL_ID = "gemini-2.5-flash"  # @param {type: "string"}
```

## Modalidades Individuales

### Comprensión textual

Gemini puede analizar preguntas textuales y retener ese contexto a través de los siguientes prompts.

```python
question = "¿Cuál es el clima promedio en Mountain View, CA, a mediados de mayo?"
prompt = """
Considerando el clima, por favor proporciona algunas sugerencias de vestimenta.

Da ejemplos para el día y la noche.
"""

contents = [question, prompt]
response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```

### Resumen de documentos

Puedes usar Gemini para procesar documentos PDF, analizar contenido, retener información y proporcionar respuestas a consultas sobre los documentos.

El ejemplo de documento PDF utilizado aquí es el [artículo de Gemini 2.5](https://arxiv.org/pdf/2507.06261).

```python
pdf_file_uri = "gs://cloud-training/gsp1210/2507.06261v5.pdf"
pdf_file = Part.from_uri(file_uri=pdf_file_uri, mime_type="application/pdf")

prompt = "¿Cuántos tokens puede procesar el modelo?"

contents = [pdf_file, prompt]

response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```

```python
prompt = """
  Eres un especialista profesional en resumen de documentos.
  Por favor, resume el documento proporcionado.
"""

contents = [pdf_file, prompt]

response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```

### Comprensión de imágenes a través de múltiples imágenes

Una de las capacidades de Gemini es poder razonar a través de múltiples imágenes.

Este es un ejemplo del uso de Gemini para razonar qué gafas serían más adecuadas para una forma de cara ovalada.

```python
image_glasses1_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/glasses1.jpg"
image_glasses2_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/glasses2.jpg"

display(Image(image_glasses1_url, width=150))
display(Image(image_glasses2_url, width=150))

prompt = """
Tengo una cara ovalada. Dada la forma de mi cara, ¿qué gafas serían más adecuadas?

Explica cómo llegaste a esta decisión.
Proporciona tu recomendación basada en la forma de mi cara y, por favor, da una explicación para cada una.
"""

contents = [
    prompt,
    Part.from_uri(file_uri=image_glasses1_url, mime_type="image/jpeg"),
    Part.from_uri(file_uri=image_glasses2_url, mime_type="image/jpeg"),
]
response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```

### Generar una descripción de video

Gemini también puede extraer etiquetas a lo largo de un video:

```python
video_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/mediterraneansea.mp4"
display(Video(video_url, width=350))

prompt = """
¿Qué se muestra en este video?
¿A dónde debería ir para verlo?
¿Cuáles son los 5 mejores lugares del mundo que se parecen a este?
¿Proporciona las 10 mejores etiquetas para este video?
"""

video = Part.from_uri(
    file_uri=video_url,
    mime_type="video/mp4",
)
contents = [prompt, video]

response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```

> Puedes confirmar que la ubicación es efectivamente Antalya, Turquía, visitando la página de Wikipedia: https://en.wikipedia.org/wiki/Antalya

También puedes usar Gemini para recuperar información adicional más allá del contenido del video.

```python
video_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/ottawatrain3.mp4"
display(Video(video_url, width=350))

prompt = """
¿Qué línea de tren es esta?
¿A dónde va?
¿Cuáles son las estaciones/paradas?
¿Qué río se está cruzando?
"""

video = Part.from_uri(
    file_uri=video_url,
    mime_type="video/mp4",
)
contents = [prompt, video]

response = client.models.generate_content(
    model=MODEL_ID, contents=contents, config=GenerateContentConfig(temperature=0)
)
display(Markdown(response.text))
```

> Puedes confirmar que esta es efectivamente la Línea de la Confederación en Wikipedia aquí: https://en.wikipedia.org/wiki/Confederation_Line

### Comprensión de audio

Gemini puede procesar audio directamente para una comprensión de contexto largo.

```python
audio_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/pixel_podcast.mp3"
display(Audio(url=audio_url))

prompt = "Por favor, resume este podcast y enumera los puntos clave."

audio = Part.from_uri(
    file_uri=audio_url,
    mime_type="audio/mpeg",
)
contents = [prompt, audio]

response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```

### Análisis de una base de código

En este caso de uso, se utilizará una herramienta externa llamada [gitingest](https://gitingest.com/) para analizar el [Repositorio de Microservicios de Google Cloud](https://github.com/GoogleCloudPlatform/microservices-demo).

```python
SUMMARY, TREE, CONTENT = ingest("https://github.com/GoogleCloudPlatform/microservices-demo")

prompt = f"""
Eres un experto analista de bases de código.
A continuación se muestra el árbol de archivos y el contenido del repositorio:

Árbol de archivos:
{TREE}

Contenido del repositorio:
{CONTENT}

Ahora, por favor:
1. Explica el objetivo del repositorio.
2. Proporciona los pasos para desplegar el repositorio.
3. Identifica todos los servicios involucrados en el repositorio y sus lenguajes de programación.
"""

response = client.models.generate_content(model=MODEL_ID, contents=prompt)
display(Markdown(response.text))
```

## Combinación de Modalidades

### Recomendaciones basadas en preferencias del usuario para comercio electrónico

Gemini puede usar el historial de compras del usuario y sus intereses actuales para proporcionar recomendaciones de productos.

```python
image_product1_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/product1.jpg"
image_product2_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/product2.jpg"

display(Image(image_product1_url, width=150))
display(Image(image_product2_url, width=150))

prompt = """
A un usuario le gusta el estilo de la primera imagen y necesita una recomendación basada en ella.
Describe los artículos y estilos en la primera imagen.
Dada la segunda imagen, ¿podrías recomendar algún artículo de ella? ¿Por qué?
"""

contents = [
    prompt,
    Part.from_uri(file_uri=image_product1_url, mime_type="image/jpeg"),
    Part.from_uri(file_uri=image_product2_url, mime_type="image/jpeg"),
]
response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```

### Comprender gráficos y diagramas

Gemini puede interpretar imágenes que contienen gráficos o diagramas.

```python
image_chart_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/chart.png"
display(Image(image_chart_url, width=500))

prompt = "Explica este gráfico y resume los puntos clave."

contents = [
    prompt,
    Part.from_uri(file_uri=image_chart_url, mime_type="image/png"),
]
response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```

### Comparar imágenes para encontrar similitudes, anomalías o diferencias

Gemini puede comparar una serie de imágenes.

```python
image_anomaly1_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/anomaly1.jpg"
image_anomaly2_url = "https://storage.googleapis.com/github-repo/img/gemini/multimodality_usecases_overview/anomaly2.jpg"

display(Image(image_anomaly1_url, width=150))
display(Image(image_anomaly2_url, width=150))

prompt = "Identifica qué es anormal en la segunda imagen basada en la primera."

contents = [
    prompt,
    Part.from_uri(file_uri=image_anomaly1_url, mime_type="image/jpeg"),
    Part.from_uri(file_uri=image_anomaly2_url, mime_type="image/jpeg"),
]
response = client.models.generate_content(model=MODEL_ID, contents=contents)
display(Markdown(response.text))
```
